```python
import pandas as pd
import csv
data = pd.read_csv("train.csv")
data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y')
data['Day'] = data['Date'].dt.day
data['Month'] = data['Date'].dt.month
data['Year'] = data['Date'].dt.year
data['Weekday'] = data['Date'].dt.weekday 
```

```python
with open("trn.csv", 'w') as c:
    csv_writer = csv.writer(c)
    csv_writer.writerow(["Date","Price","Day", "Month","Year","Weekday"])
    csv_writer.writerows(data.values.tolist())
print("done")
```

```python
data = pd.read_csv("crude-oil-price.csv")
data['date'] = pd.to_datetime(data['date'], format='%Y-%m-%dT%H:%M:%S')
data['Day'] = data['date'].dt.day
data['Month'] = data['date'].dt.month
data['Year'] = data['date'].dt.year
data['Weekday'] = data['date'].dt.weekday 
with open("crd.csv", 'w') as c:
    csv_writer = csv.writer(c)
    csv_writer.writerow(["date","price","precentChange","change","Day", "Month","Year","Weekday"])
    csv_writer.writerows(data.values.tolist())
print("done")
```
